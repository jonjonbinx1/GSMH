```python
import pandas as pd
import numpy as np

df = pd.read_csv("data/aggregated_data.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Ensure 'year' is treated as a categorical variable for clearer plotting
df['YEAR'] = df['YEAR'].astype(str)

# Create the barplot with average util_rate by year
plt.figure(figsize=(10, 6))
sns.barplot(x='YEAR', y='UTIL_RATE', data=df, estimator=np.mean, ci=None)  # Use np.mean for averaging

# Customize the plot
plt.title('Average Utilization Rate by Year')
plt.xlabel('Year')
plt.ylabel('Average Utilization Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create the barplot with average util_rate by year
plt.figure(figsize=(10, 6))
sns.barplot(x='STATE', y='UTIL_RATE', data=df, estimator=np.mean, ci=None)  # Use np.mean for averaging

# Customize the plot
plt.title('Average Utilization Rate by State')
plt.xlabel('State')
plt.ylabel('Average Utilization Rate')
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Select the columns for which you want to see total counts
columns_to_count = ['ANXIETYFLG', 'TRAUSTREFLG', 'ADHDFLG', 'BIPOLARFLG', 'DEPRESSFLG']  # Replace with the actual column names you want to analyze
total_counts = df[columns_to_count].sum()

# Create a dictionary for custom labels
column_labels = {
    'ANXIETYFLG': 'Anxiety',  # Replace with descriptive label for column1
    'TRAUSTREFLG': 'PTSD',  # Replace with descriptive label for column2
    'ADHDFLG': 'ADHD',   # Replace with descriptive label for column3
    'BIPOLARFLG': 'Bipolar',   # Replace with descriptive label for column3
    'DEPRESSFLG':'Depression',   # Replace with descriptive label for column3
}

# Map the original column names to custom labels
custom_labels = [column_labels[col] for col in total_counts.index]

# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(total_counts, labels=custom_labels, autopct='%1.1f%%', startangle=140)

# Customize the chart
plt.title('Total Percentage of Mental Health Diagnosis')
plt.show()
```

```python
# Calculate frequency of each state
state_counts = df['STATE'].value_counts()

# Calculate proportion of each state
state_proportions = df['STATE'].value_counts(normalize=True)

# Combine into a summary DataFrame
state_summary = pd.DataFrame({
    'Frequency': state_counts,
    'Proportion': state_proportions
})

# Display the summary
state_summary
```

```python
summary_stats = df.describe().transpose()
summary_stats
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Plot the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=False, cmap="coolwarm", center=0)
plt.title("Correlation Matrix of Continuous Variables")
plt.show()
```
